```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
sns.set_palette("Set2")

data = pd.read_excel(r'climate change patterns.xlsx')
data = data.replace(np.nan, 0)
data
```

```python
conditions = [
    (data['min.temp.pat'] == 'down'),
    (data['min.temp.pat'] == 'up'),
    (data['min.temp.pat'] == 'steady')]
choices = ['Not sustainable', 'Sustainable', 'Quite sustainable']
data['sust.level.min'] = np.select(conditions, choices)

conditions = [
    (data['max.temp.pat'] == 'down'),
    (data['max.temp.pat'] == 'up'),
    (data['max.temp.pat'] == 'steady')]
choices = ['Not sustainable', 'Sustainable', 'Quite sustainable']
data['sust.level.max'] = np.select(conditions, choices)

conditions = [
    (data['sust.level.max'] == 'Sustainable') & (data['sust.level.min'] == 'Sustainable'),
    (data['sust.level.max'] == 'Sustainable') & (data['sust.level.min'] == 'Quite sustainable'),
    (data['sust.level.min'] == 'Sustainable') & (data['sust.level.max'] == 'Quite sustainable'),
    (data['sust.level.max'] == 'Sustainable') & (data['sust.level.min'] == 'Not sustainable'),
    (data['sust.level.min'] == 'Sustainable') & (data['sust.level.max'] == 'Not sustainable'),
    (data['sust.level.max'] == 'Quite sustainable') & (data['sust.level.min'] == 'Quite sustainable'),
    (data['sust.level.max'] == 'Quite sustainable') & (data['sust.level.min'] == 'Not sustainable'),
    (data['sust.level.max'] == 'Not sustainable') & (data['sust.level.max'] == 'Not sustainable'),
    (data['sust.level.min'] == 'Quite sustainable') & (data['sust.level.min'] == 'Not sustainable')]
choices = ['Sustainable', 'Quite sustainable', 'Quite sustainable', 
           'Quite sustainable', 'Quite sustainable', 'Quite sustainable', 
           'Not sustainable', 'Not sustainable', 'Not sustainable']
data['sust.level.clim'] = np.select(conditions, choices)

data.to_excel(r'climate_patterns.xlsx')
data
```
