```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
sns.set_palette("Set2")

data1 = pd.read_excel(r'final.xlsx')
data1 = data1.iloc[: , 2:]
```

```python
a1=data1.groupby(['country'])['urban.agricult.perc'].mean().reset_index(name='urban.agricult.perc').set_index(['country'])
a2=data1.groupby(['country'])['hdi'].mean().reset_index(name='hdi').set_index(['country'])
a3=data1.groupby(['country'])['aqi'].mean().reset_index(name='aqi').set_index(['country'])
a4=data1.groupby(['country'])['co2.em'].mean().reset_index(name='co2.em').set_index(['country'])
a = pd.merge(a1,a2, on='country')
a = pd.merge(a,a3, on='country')
a4.sort_values(by='co2.em',ascending=False)
```

```python
data = data1._get_numeric_data()
```

```python
d=data.groupby(['year'])['urban.agricult.perc'].mean().reset_index(name='urban.agricult.perc').set_index('year')
d['urban.agricult.perc'].mean()
```

```python
d=data.groupby(['year'])['hdi'].mean().reset_index(name='hdi').set_index('year')
d['hdi'].mean()
```

```python
d=data.groupby(['year'])['aqi'].mean().reset_index(name='aqi').set_index('year')
d['aqi'].mean()
```

```python
from sklearn import linear_model 
```

```python
X = data[['year', 'hdi', 'aqi', 'urban.agricult.perc']]
y = data['co2.em'] 
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y) 
```

```python
predictedCO2 = regr.predict([[2050, 0.678, 123, 44]]) 
```

```python
print("Predicted average CO2 emissions in 2050 are: " + str(predictedCO2))
```

```python
d = data.groupby(['year'])['co2.em'].mean().reset_index(name='co2.em')
new = {'year':2050, 'co2.em':860559269}
d = d.append(new, ignore_index=True)
d = d.set_index('year')

new = pd.DataFrame([new]).set_index('year')

d['co2.em'].plot()
new['co2.em'].plot(marker='o')
plt.title(f"Average CO2 emissions \n prediction for 2050", size=15)
plt.xlabel('Year')
plt.ylabel('Emissions [t]')
plt.show()
```
